### Question 4

- split the ar4 horizontally from half
- split the ar4 into 3 vertical using array split
- split the ar4 into 3 herizontally using array split
- split the ar4 horizontally from index 5 using split

```python
ar4
```

```python
# split the ar4 horizontally from half
np.hsplit(ar4, 2)
# split horizontally from half
```

```python
# split the ar4 into 3 vertical using array split
np.array_split(ar4, 3)
np.array_split(ar4, 3, axis=0)
# both are the same thing
```

```python
# split the ar4 into 3 herizontally using array split
np.array_split(ar4, 3, axis=1)
# splitting horizontally, from column
```

```python
# split the ar4 horizontally from index 5 using split
np.split(ar4, [5], axis=1)
```

```python
# transpose ar4
ar4.transpose
ar4.T
# both do the same thing
```
